### Bonus Chanllenge

```python
def analyze(path):
    df = pd.read_csv(path)
    df["Revenue"] = df["Quantity"] * df["Price"]
    total_revenue = df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
    
    best_selling = df.groupby("Product")["Quantity"].sum().sort_values(ascending=False)[:3]
    
    return total_revenue,best_selling

    
```

```python
total_revenue , best_selling = analyze("sales_data.csv")
print("Total Revenue by Products are :")
print(total_revenue)
print("Best 3 products are :")
print(best_selling)
```
